**Below function will print the missing values (%) in sorting order**

```python
# Below code finds the percentage of missing values for each column.
# Output displays the top 5 columns with their missing value percentage.

train_missing = {} #create an empty dictionary to store the column and its percentage
print('The percentage of missing values in train data : ---------------')

#Iterate through each column and find percentage of missing values in it.
for col in train.columns:
    per = (train[col].isnull().sum()/2051)*100
    #Add only columns with missing values to the dictionary
    if per != 0.0:
        train_missing[col] = per.round(2)
    #sort the dictionary by value in descending order   
    train_missing_sort = {k: v for k, v in sorted(train_missing.items(), reverse= True, key=lambda item: item[1])}

#below will show the columns with their missing percentage of values
train_missing_sort
```

### Data Cleaning : Fill "0" for numeric columns with missing values.

As the values missing means that there is no such feature available for the house.
For eg: garage_area empty value means that there is no garage in the house.

```python
#Function to fill "0" values in the columns of integer type

def fillna(columns):
    #stores the list of index where value is missing
    index = train[train[columns].isnull()].index.tolist()
    
    # Iterate through the index and fill "0" value
    for i in index:
        train[columns].iloc[i] = 0
```

```python
#List of columns to fill "0" for missing values
columns_fill0 = ['bsmtfin_sf_1','bsmtfin_sf_2','bsmt_unf_sf',
                'total_bsmt_sf','garage_cars','garage_area',
                'mas_vnr_area','bsmt_full_bath','bsmt_half_bath',
                'lot_frontage','garage_yr_blt']

# Call the fillna function to fill "0"
for columns in columns_fill0:
    fillna(columns)
```

```python
#Convert the below datatype to int instead of float

columns_to_int = ['lot_frontage','mas_vnr_area','bsmtfin_sf_1','bsmtfin_sf_2','bsmt_unf_sf','total_bsmt_sf',
                 'bsmt_full_bath','bsmt_half_bath','garage_yr_blt','garage_cars','garage_area']

for col in columns_to_int:
    train[col] = train[col].astype(int)
```

### Data Cleaning : Fill "None" for categorical columns with missing values.

As the values missing means that there is no such feature available for the house. <br>
For eg: garage_type empty value means that there is no garage in the house, therefore, there will be no garage_type.

```python
#List of columns to fill "None" for missing values in columns with datatype: Object
columns_fillNone = ['mas_vnr_type','bsmt_qual','bsmt_cond','bsmt_exposure',
                   'bsmtfin_type_1','bsmtfin_type_2','garage_type','garage_finish',
                   'garage_qual','garage_cond','fireplace_qu']

# Iterate through the columns in list and fill "None" for missing values
for col in columns_fillNone:
    train[col].fillna('None',inplace=True)
```

For example: We can assign the values like excellent, good, average, fair, poor to numeric values as below:

- Good : 4

```python
# Dictionary for ordinal columns
# Keys in dictionary contain the column names
# Values of each key is a dictionary with assignment details.
dict_columns_binarize = {'central_air': {'Y':1, 'N':0},
                        'exter_qual': {'Ex':5, 'Gd':4, 'TA': 3, 'Fa':2},
                        'exter_cond': {'Ex':5, 'Gd':4, 'TA': 3, 'Fa':2, 'Po':1},
                        'bsmt_qual': {'Ex':5, 'Gd':4, 'TA': 3, 'Fa':2, 'Po':1, 'None':0},
                        'bsmt_cond': {'Ex':5, 'Gd':4, 'TA': 3, 'Fa':2, 'Po':1, 'None':0},
                        'heating_qc': {'Ex':5, 'Gd':4, 'TA': 3, 'Fa':2, 'Po':1},
                        'kitchen_qual' : {'Ex':5, 'Gd':4, 'TA': 3, 'Fa':2},
                        'fireplace_qu': {'Ex':5, 'Gd':4, 'TA': 3, 'Fa':2, 'Po':1, 'None':0},
                        'garage_qual': {'Ex':5, 'Gd':4, 'TA': 3, 'Fa':2, 'Po':1, 'None':0},
                        'garage_cond': {'Ex':5, 'Gd':4, 'TA': 3, 'Fa':2, 'Po':1, 'None':0},
                        'bsmt_exposure': {'Gd':3, 'Av': 2, 'Mn':1, 'None':0, 'No':0},
                        'bsmtfin_type_1' : {'GLQ':4 ,'ALQ':3 ,'BLQ':2 ,'Rec':3, 'LwQ':1 ,'Unf':0 ,'None': 0},
                        'bsmtfin_type_2' : {'GLQ':4 ,'ALQ':3 ,'BLQ':2 ,'Rec':3, 'LwQ':1 ,'Unf':0 ,'None': 0},
                        'lot_shape' : {'Reg':4, 'IR2':3, 'IR3':2, 'IR1':1},
                        'utilities' : {'AllPub':3, 'NoSeWa':2, 'NoSewr':1},
                        'land_slope' : {'Gtl':3, 'Mod':2, 'Sev':1},
                        'electrical' : {'SBrkr':5, 'FuseA':4 ,'FuseF':3, 'FuseP':2 ,'Mix':1},
                        'functional' : {'Typ':8, 'Min1': 7, 'Min2':6, 'Mod':5, 'Maj1':4,'Maj2':3,'Sev':2, 'Sal':1},
                        'garage_finish' : {'Fin':3, 'RFn':2, 'Unf': 1, 'None':0},
                        'paved_drive' : {'Y':2,'P':1, 'N':0}
                        } 
```

```python
# List of ordinal columns to be converted to numeric
columns = ['central_air','exter_qual','exter_cond',
                        'bsmt_qual','bsmt_cond','heating_qc',
                        'kitchen_qual','fireplace_qu','garage_qual',
                        'garage_cond','bsmt_exposure',
                        'bsmtfin_type_1','bsmtfin_type_2','lot_shape',
                        'utilities','land_slope','electrical',
                        'functional','garage_finish','paved_drive']

# Function to convert the ordinal columns to numeric
def ordinal_columns_to_int(col):
    return  train[col].map(dict_columns_binarize[col])
```

```python
#List of columns to fill "0" for missing values
columns_fill0 = ['lot_frontage','mas_vnr_area','garage_yr_blt']

# Call the fillna function to fill "0"
for columns in columns_fill0:
    fillna(columns)
```
